# Ronaldo's Stats Season wise

### Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Variables

```
seasons=list(range(2002,2020))
appearances=[25,29,33,33,34,34,33,29,34,38,34,30,35,36,29,27,31,33]
goals=[3,4,5,9,17,31,18,26,40,46,34,31,48,35,25,26,21,31]
assists=[3,7,6,6,13,6,6,7,9,12,10,9,16,11,6,5,8,5]
```

### Making data frame of the variables

```
Ronaldo_dict={"Seasons":seasons, "Appearances":appearances, "Goals":goals, "Assists":assists}
```

```
Ronaldo_dict
```

```
ronaldo_df=pd.DataFrame(Ronaldo_dict)
```

```
ronaldo_df.Seasons = ronaldo_df.Seasons.astype(str)
```

```
print(ronaldo_df)
```

### Plotting Columns against Seasons

```
plt.figure(figsize=(16,10))
plt.plot(ronaldo_df.Seasons,ronaldo_df.Appearances, label="Appearances")
plt.plot(ronaldo_df.Seasons,ronaldo_df.Goals, label="Goals")
plt.plot(ronaldo_df.Seasons,ronaldo_df.Assists, label="Assists")
plt.title("Appearances and Goals")
plt.xlabel("Seasons")
plt.legend()
```

### Making Modified Fields and Plotting Them

```
goal_per_appearance = ronaldo_df.Goals/ronaldo_df.Appearances
assist_per_appearance = ronaldo_df.Assists/ronaldo_df.Appearances
total_goal_contribution = ronaldo_df.Goals+ronaldo_df.Assists
```

```
plt.figure(figsize=(16,10))
plt.plot(ronaldo_df.Seasons, goal_per_appearance, label="Goals/Appearance")
plt.title("Goals per Appearance")
plt.xlabel("Seasons")
plt.legend()
```

```
plt.figure(figsize=(16,10))
plt.plot(ronaldo_df.Seasons, assist_per_appearance, label="Assists/Appearance")
plt.title("Assists per Appearance")
plt.xlabel("Seasons")
plt.legend()
```

```
plt.figure(figsize=(16,10))
plt.plot(ronaldo_df.Seasons, total_goal_contribution, label="Goal Contribution")
plt.title("Net Goal contibution")
plt.xlabel("Seasons")
plt.legend()
```
