# Messi in la liga 

```
# Attributes

seasons=list(range(5,20))
appearances=[17,26,28,31,35,33,37,32,31,38,33,34,36,34,33]
goals=[6,14,10,23,34,31,50,46,28,43,26,37,34,36,25]
assists=[2,2,12,11,10,18,16,12,11,18,16,9,12,13,21]
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
messi_dict={"Seasons":seasons, "Appearances":appearances,"Goals":goals,"Assists":assists}
```

```
messi_dict
```

```
messi_df=pd.DataFrame(messi_dict)
```

```
print(messi_df)
```

```
plt.plot(messi_df.Seasons,messi_df.Appearances, label="Appearances")
plt.plot(messi_df.Seasons,messi_df.Goals, label="Goals")
plt.title("Score vs Seasons")
plt.ylabel("Score")
plt.xlabel("Seasons")
plt.xlim(5, 19)
plt.legend()
```

```
messi_df.Goals/messi_df.Appearances
```

```
plt.plot(messi_df.Seasons, messi_df.Goals/messi_df.Appearances, label="Goal/Appearances")
```

```
plt.plot(messi_df.Seasons, messi_df.Assists/messi_df.Appearances, label="Assists/Appearances")
```
