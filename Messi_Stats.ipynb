# Messi in la liga 

### Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Variables

```
seasons=list(range(2005,2020))
appearances=[17,26,28,31,35,33,37,32,31,38,33,34,36,34,33]
goals=[6,14,10,23,34,31,50,46,28,43,26,37,34,36,25]
assists=[2,2,12,11,10,18,16,12,11,18,16,9,12,13,21]
```

### Making data frame of the variables

```
messi_dict={"Seasons":seasons, "Appearances":appearances,"Goals":goals,"Assists":assists}
```

```
messi_dict
```

```
messi_df=pd.DataFrame(messi_dict)
```

```
messi_df.Seasons = messi_df.Seasons.astype(str)
```

```
print(messi_df)
```

### Plotting Columns against Seasons

```
plt.figure(figsize=(16,10))
plt.plot(messi_df.Seasons,messi_df.Appearances, label="Appearances")
plt.plot(messi_df.Seasons,messi_df.Goals, label="Goals")
plt.plot(messi_df.Seasons,messi_df.Assists, label="Assists")
plt.title("Appearances and Goals")
plt.xlabel("Seasons")
plt.legend()
```

### Making Modified Fields and Plotting Them

```
goal_per_appearance = messi_df.Goals/messi_df.Appearances
assist_per_appearance = messi_df.Assists/messi_df.Appearances
total_goal_contribution = messi_df.Goals+messi_df.Assists
```

```
plt.figure(figsize=(16,10))
plt.plot(messi_df.Seasons, goal_per_appearance, label="Goals/Appearance")
plt.title("Goals per Appearance")
plt.xlabel("Seasons")
plt.legend()
```

```
plt.figure(figsize=(16,10))
plt.plot(messi_df.Seasons, assist_per_appearance, label="Assists/Appearance")
plt.title("Assists per Appearance")
plt.xlabel("Seasons")
plt.legend()
```

```
plt.figure(figsize=(16,10))
plt.plot(messi_df.Seasons, total_goal_contribution, label="Goal Contribution")
plt.title("Net Goal contibution")
plt.xlabel("Seasons")
plt.legend()
```
